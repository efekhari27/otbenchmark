# Marginal and conditional distributions

## Conditional distribution

For a given pair of continuous random variables $X$ and $Y$ with probability densities $f_X$ and $f_Y$, the conditional density of $Y|X$ is:

$$
f_{Y|X}(y|X = x) = \frac{f_{X,Y}(x,y)}{f_X(x)},
$$

where $f_X(x)$ is the marginal distribution of $X$:

$$
f_X(x) = \int_{\mathbb{R}^{|Y|}} f_{X,Y}(x,y) dy.
$$

In the previous equation, the integer $|Y|$ is the dimension of the vector $Y$. 

For a multivariate random variable $X=(X_1, X_2, ..., X_d)$ and we consider conditional distribution which arise when we condition with respect with several marginals of $X$. 
More precisely, we assume that a given set of conditioning indices 

$$
C=(i_1, i_2, ..., i_c)
$$

are given, where $0\leq c\leq d$ is an integeger, $i_1, i_2, ..., i_c \in \{1, 2, ..., d\}$ and are so that $i_1 < i_2 < ... <i_c$. In other words, the integers $i_1, ..., i_c$ are unique indices of $X$. We also assume that we are given the corresponding set of conditioning values 

$$
x_C = (x_{i_1}, x_{i_2}, ..., x_{i_c}) \in\mathbb{R}^c.
$$

We denote by $\overline{C}$ the complementary set of $C$:
$$
\overline{C} = \{i \in \{1, ..., d\} | i \not\in C\}.
$$
Therefore, the sets $C$ and $\overline{C}$ partition the indices of $X$, so that:

$$
X = \left(X_C, X_\overline{C}\right).
$$

The conditional distribution of $X$ given the conditioning indices $C$ and the conditioning values $x_C$ is the distribution of the random variable $X$ given the conditioning marginals have been set to $x_C$.
Its probability density function is:

$$
f_{X|X_C=x_C}(x_\overline{C})= \frac{f_X(x_C, x_\overline{C})}{f_{X_C}(x_C)}
$$

where $f_{X_C}(x_C)$ is the marginal probability density function of $X_C$:

$$
f_{X_C}(x_C) = \int_{\mathbb{R}^{d - c}} f_\left(X_C, X_\overline{C}\right)\left(x_C, x_\overline{C}\right) dx_\overline{C}.
$$

## Example of conditional distributions in three dimensions

For example, consider the random variable $X = (X_1, X_2, X_3)$ with probability 
density function $f_X$. 
The conditional distribution of $(X_1, X_2, X_3 | X_2 = x_2, X_3 = x_3)$ is:

$$
f_{(X_1, X_2, X_3 | X_2 = x_2, X_3 = x_3)}(x_1) 
= \frac{f_X(x)}{f_{(X_2, X_3)}(x_2, x_3)}
$$

where $f_{(X_2, X_3)}$ is the marginal distribution of $(X_2,X_3)$:

$$
f_{(X_2, X_3)}(x_2, x_3) = \int_{\mathbb{R}} f_X(x_1, x_2, x_3) dx_1.
$$

The conditional distribution of $(X_1, X_2, X_3 | X_3 = x_3)$ is:

$$
f_{(X_1, X_2, X_3 | X_3 = x_3)}(x_1, x_2) 
= \frac{f_X(x)}{f_{X_3}(x_3)}
$$

where $f_{X_3}$ is the marginal density of $X_3$:

$$
f_{X_3}(x_3) = \int_{\mathbb{R}^2} f_X(x_1, x_2, x_3) dx_1 dx_2.
$$
